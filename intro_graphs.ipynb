```python
import matplotlib.pyplot as plt

compounds = ['Soft', 'Medium', 'Hard']
speed = [3, 2, 1]
durability = [1, 2, 3]
colors = ['#c5211c', '#cfb93f', '#fefefe']

bars = []
for i, compound in enumerate(compounds):
    bar = plt.bar(durability[i], speed[i], color=colors[i], edgecolor='black', label=compound)
    bars.append(bar)

plt.xlabel('Durability')
plt.ylabel('Speed / Grip')
plt.title('Tyre Compounds: Speed vs Durability')

plt.xticks(range(min(durability), max(durability)+1))
plt.yticks(range(min(speed), max(speed)+1))

plt.legend(title="Tyre Compound", loc="upper right")

plt.tight_layout()
plt.show()
```
